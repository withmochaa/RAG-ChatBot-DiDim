```python
import pandas as pd

# 엑셀 파일 경로
excel_file_path = '/Users/withmocha/Desktop/2024-2강의시간표(한국어)_20240809.xlsx'  

# 엑셀 파일 읽기
df = pd.read_excel(excel_file_path)

# 열 이름에서 '\n' 문자 제거
df.columns = df.columns.str.replace('\n', '', regex=False)

# 각 열마다 데이터프레임 만들기
dataframes = {}  # 각 열의 데이터프레임을 저장할 딕셔너리

for column in df.columns:
    # 현재 열의 데이터를 포함하는 데이터프레임 생성
    dataframes[column] = pd.DataFrame(df[column])
    
    # NaN 값을 'X'로 변경
    dataframes[column].fillna('X', inplace=True)

    # 데이터프레임을 변수로도 저장
    globals()[column] = dataframes[column]  # 열 이름을 변수 이름으로 사용

# 결과 출력
for column, dataframe in dataframes.items():
    print(f"\n'{column}")
    print(dataframe.head(10) if len(dataframe) > 10 else dataframe)  # 10개 데이터 출력

```
